# TokenGT Attention Distance Plot

```python
import torch
import matplotlib.pyplot as plt
import numpy as np
```

```python
# TokenGT Laplacian Embedding
attn = torch.load('/main/tokengt/large-scale-regression/scripts/ckpts/pcqv2-tokengt-lap16-trained/checkpoint_best_visualize_valid.pt')  # [L, H]
L, H = attn.shape # L = 12, H = 16
attn_flatten = torch.cat([attn[i,:] for i in range(L)]).cpu().numpy()  #[L*H] 
pos = []
for i in range(L): 
    pos.extend([i for _ in range(H)])
colors = [i for i in range(H)] * L 

fig, axs = plt.subplots(1, 2, figsize=(16, 8))
plt.suptitle("TokenGT Laplacian", fontsize=20)

axs[0].scatter(pos, attn_flatten, s=50, c=np.array(colors))
axs[0].set_title("avg attention distance", fontdict={'fontsize':15})
axs[0].set_xlabel('Network depth(layer)')
axs[0].set_ylabel('distance')

head_idx = [i for i in range(H)]
axs[1].scatter(np.array(head_idx), np.ones(H), s=50, c=np.array(head_idx))
axs[1].set_title("color map", fontdict={'fontsize':15})
axs[1].set_xlabel('head index')
axs[1].set_yticks([])

plt.show()
plt.close()
```
